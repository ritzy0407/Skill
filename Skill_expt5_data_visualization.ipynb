```python
import pandas as pd
df = pd.read_csv('/content/diabetes.csv')
print(df.head())
```

```python
df.dropna(inplace=True)
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['Glucose'],bins=10,kde=True)
plt.title('Histogram of Glucose')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Glucose'], df['BMI'], alpha=0.6)
plt.title("Scatter plot of Glucose vs BMI")
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns  # Get numeric columns
df[numeric_columns].hist(bins=15, figsize=(15, 10))  # Plot histograms
plt.suptitle("Histograms of Numeric Columns")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[numeric_columns])  # Boxplot for all numeric columns
plt.title("Boxplot of Numeric Columns")
plt.xticks(rotation=45)  # Rotate the column labels for better readability
plt.show()
```
